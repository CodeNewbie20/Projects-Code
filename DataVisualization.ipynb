```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Step 1: Create a sample dataset
data = {
 "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
 "City": ["New York", "Paris", "New York", "Berlin", "Paris"],
 "Gender": ["Female", "Male", "Male", "Male", "Female"],
 "Age": [24, 27, 22, 32, 29],
 "Preference": ["Sports", "Music", "Sports", "Movies", "Music"]
}

# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# -- Step 2: One-Hot encoding using pandas.get_dummies ---
print("\nOne-Hot Encodingwith Pandas (get_dummies):")
encoded_df_dummies = pd.get_dummies(df, columns=["City","Gender","Preference"], drop_first=True)
encoded_df_dummies.head()
```

```python
# Define the OneHotEncoder with the updated parameter
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop="first", sparse_output=False)

# Select categorical columns for encoding
categorical_columns = ["City", "Gender", "Preference"]

# Apply encoding and convert the result to a DataFrame
encoded_array = encoder.fit_transform(df[categorical_columns])
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_columns))

# Combine the one-hot-encoded columns with the rest of the dataset
df_encoded = pd.concat([df.drop(categorical_columns, axis=1), encoded_df], axis=1)

# Display the final dataset
print("One-Hot Encoded Dataset using sklearn's OneHotEncoder:")
print(df_encoded)
```
